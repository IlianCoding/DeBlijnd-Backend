```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel

# Load from Hugging Face
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-printed")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-printed")

# Save in a local directory
processor.save_pretrained("./model/local_trocr")
model.save_pretrained("./model/local_trocr", safe_serialization=False)  # Forces saving as pytorch_model.bin
```

```python
loaded_processor = TrOCRProcessor.from_pretrained("./model/local_trocr")
loaded_model = VisionEncoderDecoderModel.from_pretrained("./model/local_trocr")
```
